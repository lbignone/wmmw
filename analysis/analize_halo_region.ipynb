# Check the number of neighbours as a possible indicatior of enviroment

```python
dx = 0.5

neighbours = []

for i in range(halo_candidates.ID.size):
    x0 = halo_candidates.iloc[i].X
    y0 = halo_candidates.iloc[i].Y
    z0 = halo_candidates.iloc[i].Z
    
    id0 = halo_candidates.iloc[i].ID
    
    d = np.sqrt((rock.X-x0)**2 + (rock.Y-y0)**2 + (rock.Z-z0)**2)
    neighbours.append(d[(d<=0.5) & (d>0.0)].size)
    
neighbours = np.array(neighbours)    
```

```python
select_n = 3
nlim = [25, 50, 100]
labels = ["n <= 25", "25 < n <= 50", "50 < n <= 100"]

random_ids = {}

for i, nl in enumerate(nlim):
    random_ids[nl] = []
    count = 0
    while (count < select_n):
        hid = randint(0, neighbours.size-1)
        n = neighbours[hid]
        if i == 0:
            if n <= nl:
                random_ids[nl].append(hid)
                count += 1
        else:
            if (n <= nl) and (n > nlim[i-1]):
                random_ids[nl].append(hid)
                count += 1    
```
