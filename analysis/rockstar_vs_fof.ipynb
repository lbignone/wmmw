```python
basedir = "/home/lbignone/simulations/wmmw/out512"
snap_number = 10
dm_mass = 8.20612e+07
```

```python
rock = rockstar_load(basedir, snap_number)
```

```python
fof = Fof(basedir, snap_number)
fof.groupmass = fof.grouplen * dm_mass

fof.groupmass.shape = (fof.groupmass.size, 1)
data = np.concatenate([fof.groupcm/1000.0, fof.groupmass], axis=1)

df_fof = pd.DataFrame(data, columns=["X", "Y", "Z", "M"])
```

```python
id1 = 32886
id2 = 32860

center = (rock.loc[id1, ["X", "Y", "Z"]]  + rock.loc[id2, ["X", "Y", "Z"]])/2.0
```

```python
dx = 1
xmin = center.X - dx
xmax = center.X + dx

ymin = center.Y - dx
ymax = center.Y + dx

zmin = center.Z - dx
zmax = center.Z + dx

region = rock[(rock.X > xmin) & (rock.X < xmax) & (rock.Y > ymin) & (rock.Y < ymax) & (rock.Z > zmin) & (rock.Z < zmax)]
region_fof = df_fof[(df_fof.X > xmin) & (df_fof.X < xmax) & (df_fof.Y > ymin) & (df_fof.Y < ymax) & (df_fof.Z > zmin) & (df_fof.Z < zmax)]
```

```python
fig = plt.figure(figsize=(18,8))
ax = fig.add_subplot(1, 2, 1, projection='3d')

# scale size to mass
minscale = 10
maxscale = 200
a = (maxscale - minscale)/(region.M200c.max() - region.M200c.min())
b = 1 - a*region.M200c.min()
scale = a*region.M200c + b
#scale = 7*region.Rvir


# Rockstar
p = ax.scatter(region.X, region.Y, region.Z, s=scale)
#p = ax.scatter(h1.X, h1.Y, h1.Z, s=(a*h1.M200c+b), color="r")
#p = ax.scatter(h2.X, h2.Y, h2.Z, s=(a*h1.M200c+b), color="r")
ax.set_title("Rockstar")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")


# FOF
a = (maxscale - minscale)/(region_fof.M.max() - region_fof.M.min())
b = 1 - a*region_fof.M.min()
scale = a*region_fof.M + b

ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.scatter(region_fof.X, region_fof.Y, region_fof.Z, s=scale)
ax.set_title("FOF")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```
